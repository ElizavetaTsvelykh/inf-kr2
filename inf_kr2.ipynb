```python
import numpy as np
positive_sum=0
positive_count=0
matrix = np.random.randint(-20, 20, size=(6, 8))
max_element=max(max(row)for row in matrix)
print("Исходная матрица:\n", matrix)
print("Максимальный элемент:",max_element)
for element in matrix.flatten():
    if element > 0:
        positive_sum += element
        positive_count += 1
        average = positive_sum/positive_count
matrix[matrix==max_element]=average
print("Обработанная матрица:\n", matrix)
```
